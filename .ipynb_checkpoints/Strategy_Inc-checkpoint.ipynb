```python
## We are going to create a Dataframe with key financial data 

import glob
from bs4 import BeautifulSoup
import pandas as pd

# Helper functions
def get_nested(soup, keys):
    for key in keys:
        if soup is None:
            return None
        soup = soup.find(key.lower())  # html.parser lowercases all tags
    return soup.text.strip() if soup else None

def to_float(x):
    try:
        return float(x.replace(",", "")) if x else None
    except:
        return None

# Get all .txt files recursively
paths = glob.glob("sec-edgar-filings/0001079782/4/**/*.txt", recursive=True)
print(f"Found {len(paths)} .txt files")

all_transactions = []

for file_path in paths:
    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
        text = f.read()
    
    # Extract <ownershipDocument> XML block
    start = text.find("<ownershipDocument>")
    end = text.find("</ownershipDocument>")
    if start == -1 or end == -1:
        continue
    xml_block = text[start:end + len("</ownershipDocument>")]

    # Parse XML
    xml_soup = BeautifulSoup(xml_block, "html.parser")

    # Extract ticker symbol once per filing
    ticker_tag = xml_soup.find("issuertradingsymbol")
    ticker = ticker_tag.text.strip() if ticker_tag else "MSTR"

    # Capture all footnotes in this document,
    all_notes = []
    footnotes_block = xml_soup.find("footnotes")
    if footnotes_block:
        for f in footnotes_block.find_all("footnote"):
            all_notes.append(f"{f.get('id', '')}: {f.get_text(strip=True)}")
    doc_note_text = " ".join(all_notes) if all_notes else None

    # Get all transactions (non-derivative and derivative)
    tx_nodes = xml_soup.find_all(["nonderivativetransaction", "derivativetransaction"])

    for tx in tx_nodes:
        tx_dict = {
            "date": get_nested(tx, ["transactionDate", "value"]),
            "ticker": ticker,
            "code": get_nested(tx, ["transactionCoding", "transactionCode"]),
            "shares": to_float(get_nested(tx, ["transactionAmounts", "transactionShares", "value"])),
            "price": to_float(get_nested(tx, ["transactionAmounts", "transactionPricePerShare", "value"])),
            "acquire_dispose": get_nested(tx, ["transactionAmounts", "transactionAcquiredDisposedCode", "value"]),
            "post_transaction_shares": to_float(get_nested(tx, ["postTransactionAmounts", "sharesOwnedFollowingTransaction", "value"])),
            "ownership": get_nested(tx, ["ownershipNature", "directOrIndirectOwnership", "value"]),
            "class_type": get_nested(tx, ["securityTitle", "value"]),
            "file_path": file_path,
            "note_text": doc_note_text
        }
        all_transactions.append(tx_dict)

# Convert to DataFrame
df = pd.DataFrame(all_transactions)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['shares'] = pd.to_numeric(df['shares'], errors='coerce')
df['year'] = df['date'].dt.year

df
```

```python
### Now, let's use text analysis to scan footnotes to surface the most common meaningful words and tally key phrases broken out by share class

import spacy
from collections import Counter
import pandas as pd

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Focused phrases to track in footnotes
key_phrases = ["trading plan", "conversion", "vesting", "weighted average price"]

# Take a subset of 500 footnotes for testing
notes_subset = df['note_text'].dropna().head(500)

# Function to extract top words and key phrases
def analyze_footnotes(text_series, top_n=20):
    all_tokens = []
    phrase_counts = {phrase: 0 for phrase in key_phrases}
    
    for doc in nlp.pipe(text_series, batch_size=50):
        tokens = [token.lemma_.lower() for token in doc if token.is_alpha and not token.is_stop]
        all_tokens.extend(tokens)
        
        text_lower = doc.text.lower()
        for phrase in key_phrases:
            phrase_counts[phrase] += text_lower.count(phrase)
    
    top_words = Counter(all_tokens).most_common(top_n)
    return top_words, phrase_counts

# Analyze the subset
top_words, phrase_counts = analyze_footnotes(notes_subset)

print("Top words in subset of footnotes:", top_words)
print("Key phrase counts in subset:", phrase_counts)

# Optional: per class (for the subset)
for cls in df['class_type'].unique():
    cls_notes = df[df['class_type'] == cls]['note_text'].dropna().head(100)  # smaller subset per class
    cls_top_words, cls_phrase_counts = analyze_footnotes(cls_notes)
    print(f"\nClass: {cls}")
    print("Top words:", cls_top_words)
    print("Key phrase counts:", cls_phrase_counts)
```

```python
# Let's analyze Class A stock transactions

import pandas as pd
import numpy as np

# Filter for Class A Common Stock and clean data
class_a_tx = df[df['class_type'] == 'Class A Common Stock'].copy()
class_a_tx['date'] = pd.to_datetime(class_a_tx['date'], errors='coerce')
class_a_tx['shares'] = pd.to_numeric(class_a_tx['shares'], errors='coerce')
class_a_tx = class_a_tx.dropna(subset=['date', 'shares'])

# Extract year
class_a_tx['year'] = class_a_tx['date'].dt.year

# Compute acquired vs disposed shares
class_a_tx['shares_acquired'] = np.where(class_a_tx['acquire_dispose'] == 'A', class_a_tx['shares'], 0)
class_a_tx['shares_disposed'] = np.where(class_a_tx['acquire_dispose'] == 'D', class_a_tx['shares'], 0)

# Aggregate by year
annual_class_a = (
    class_a_tx
    .groupby('year', as_index=False)
    .agg(
        total_shares_acquired=('shares_acquired', 'sum'),
        total_shares_disposed=('shares_disposed', 'sum')
    )
)

# Compute net change
annual_class_a['net_shares_change'] = annual_class_a['total_shares_acquired'] - annual_class_a['total_shares_disposed']

# Display
print("Annual Class A Stock Transactions")
print(annual_class_a)
```

```python
# Let's analyze Class B stock transactions 

import pandas as pd
import numpy as np

# Filter for Class B Common Stock and clean data
class_b_tx = df[df['class_type'] == 'Class B Common Stock'].copy()
class_b_tx['date'] = pd.to_datetime(class_b_tx['date'], errors='coerce')
class_b_tx['shares'] = pd.to_numeric(class_b_tx['shares'], errors='coerce')
class_b_tx = class_b_tx.dropna(subset=['date', 'shares'])

# Extract year
class_b_tx['year'] = class_b_tx['date'].dt.year

# Compute acquired vs disposed shares
class_b_tx['shares_acquired'] = np.where(class_b_tx['acquire_dispose'] == 'A', class_b_tx['shares'], 0)
class_b_tx['shares_disposed'] = np.where(class_b_tx['acquire_dispose'] == 'D', class_b_tx['shares'], 0)

# Aggregate by year
annual_class_b = (
    class_b_tx
    .groupby('year', as_index=False)
    .agg(
        total_shares_acquired=('shares_acquired', 'sum'),
        total_shares_disposed=('shares_disposed', 'sum')
    )
)

# Compute net change
annual_class_b['net_shares_change'] = annual_class_b['total_shares_acquired'] - annual_class_b['total_shares_disposed']

# Display
print("Annual Class B Stock Transactions")
print(annual_class_b)
```

```python
# I used AI to help me visualize this better 
import matplotlib.pyplot as plt

# Merge Class A and Class B annual data for easy plotting
annual_a = annual_class_a[['year', 'net_shares_change']].rename(columns={'net_shares_change': 'Class A'})
annual_b = annual_class_b[['year', 'net_shares_change']].rename(columns={'net_shares_change': 'Class B'})

annual_tx = pd.merge(annual_a, annual_b, on='year', how='outer').sort_values('year').fillna(0)

# Plot
plt.figure(figsize=(12,6))
plt.plot(annual_tx['year'], annual_tx['Class A'], marker='o', label='Class A Net Shares Change')
plt.plot(annual_tx['year'], annual_tx['Class B'], marker='s', label='Class B Net Shares Change')

# Highlight 2020
plt.axvline(x=2020, color='orange', linestyle='--', linewidth=2, label='Bitcoin Pivot Year')
plt.text(2020, max(annual_tx['Class A'].max(), abs(annual_tx['Class B'].min()))*0.8, '2020: Bitcoin Pivot', 
         rotation=90, color='orange', fontsize=12, verticalalignment='center')

plt.title('MicroStrategy Class A vs Class B Net Stock Transactions by Year')
plt.xlabel('Year')
plt.ylabel('Net Shares Change')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Conclusion: We aim to track Michael Saylor’s insider trades to see whether he has been rapidly selling off Class A stock while maintaining control via Class B shares, thereby managing his personal exposure without relinquishing voting power. Additionally, we analyzed footnotes to understand the language and mechanisms associated with Class A versus Class B transactions.

Findings:

Class A shares show relatively stable net changes. In 2024, heavy sales of Class A shares were offset by acquisitions via option exercises, leaving net changes close to zero. Public filings confirm these moves were part of Rule 10b5‑1 trading plans, suggesting structured management rather than reactive selling.

Class B shares display a “see-saw” pattern: large net disposals in early years (2005–2012) and minimal activity afterward. Most Class B movements reflect conversions to Class A or internal restructuring rather than public market activity. Footnote analysis shows “trading plan” is frequently mentioned in Class A notes but rarely in Class B notes, and key terms like option exercise, weighted average price, and conversion are more prominent for Class A.


Interpretation Relative to Hypothesis:

Saylor has been selling Class A shares, but this is likely tied to a regular cycle of trading plans. The footnotes reinforce that these trades are planned and strategic, not based on reactions to Bitcoin's changing fortunes, so our hypothesis seems to be disproven. However, he preserves control via Class B shares, and that's where it seems the more interesting story lies. 

Probing Questions / Investigative Angles:

We see an increase in the disposal of Class B shares in 2020 and huge disposals of Class B from 2005–2012. Who received those Class B shares - private investors, employees, or related parties? Is there an unusual concentration of ownership or transfers to related parties?

The stock that confers control, Class B, became completely immobile right after the company entered its most volatile era in 2020. No Class B grants to employees is atypical for a tech-leaning company and a sign that Saylor may be consolidating control.
